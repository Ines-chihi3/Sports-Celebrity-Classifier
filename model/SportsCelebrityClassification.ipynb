#### read image

```python
#read image
img = cv2.imread('./test_images/sharapova1.jpg')
#convert image  from RGB to gray 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

#### drow rectangle on face and eye using haar cascade

```python
#create face casscade ( all haarcascades are on this directory in my local host)
#i use 'haarcascade_frontalface_default.xml' for face detection and 'haarcascade_eye.xml' for eye detection
face_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')
#detect face location
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
#face_cascade.detectMultiScale :  return (x,y,w,h)  for face location 
```

```python
#drow rectangle on face and eye
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
#eye_cascade.detectMultiScale :  return (ex,ey,ew,eh)  for eye location
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        

plt.figure()
plt.imshow(face_img)
plt.show()
```

### <font color=blue>apply the fonction of creating cropped image on all our dataset</font>

###### we need cropped image : because cropped image contain just face and eye all we need to classify image

##### we will crop face from all image in our dataset and we will store it in folder

```python
#fuction : input : image from dataset , output : cropped image (if face and eyes >=2 are detected)
#this fucntion retun a cropped image if the original image contain face and 2 eye 
#our goal is to safe all cropped images (face) on data set because this is the region of interest 
```

```python
def cropped_image_with_2_eyes(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

###### let's apply this fonction on our dataset

```python
#indicate the path for dataset and the path where we will save cropped images
#it will create a folder in my dataset contain all cropped image
path_to_data = "./dataset/"
path_to_crop_data = "./dataset/cropped/"
```

```python
#store path of all sub folder in python list 
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)

img_dirs
```

```python
#create folder for cropped image
os.mkdir(path_to_crop_data)
```

```python
#we ill eterate for all image and create cropped image and save it on cropped data folder
```

```python
cropped_image_dirs = []
celebrity_file_names_dict = {}

for img_dir in img_dirs:
    count = 1
    #just to extract celebrety name from './dataset/lionel_messi'
    celebrity_name = img_dir.split('/')[-1]
    celebrity_file_names_dict[celebrity_name] = []
    #iterate for all original image 
    for entry in os.scandir(img_dir):
            image = cv2.imread(entry.path);
            if image is None  : # // image was not created 
                continue
            else :
                #print(entry.path)
                #create cropped image for original image
                roi_color = cropped_image_with_2_eyes(entry.path)
                #verify if it's not none ( function return cropped image)
                if roi_color is not None:
                    #save image on cropped folder 
                    #cropped folder name
                    cropped_folder = path_to_crop_data + celebrity_name
                    #test if folder dont exist ! create it 
                    if not os.path.exists(cropped_folder):
                        os.makedirs(cropped_folder)

                        cropped_image_dirs.append(cropped_folder)
                    #we must give a name for cropped image its just 'celecrety name '+1/2/3 depending on index of imaaeg ( the first or the second..)
                    #example : messi1.png , messi2.png , messi3.png
                    cropped_file_name = celebrity_name + str(count) + ".png"
                    cropped_file_path = cropped_folder + "/" + cropped_file_name 
                    #save cropped image (roi_color ) on cropped_file_path that we create
                    cv2.imwrite(cropped_file_path, roi_color)
                    #we must store all cropped_file_path because it will help us
                    celebrity_file_names_dict[celebrity_name].append(cropped_file_path)
                    count += 1
             
```

# <font color=green>Preprocessing : Feature Engineering : Use wavelet transform</font>

##### Now we will use cropped images and apply wavelet transform to extract meaning features that can help with image identification. 

###### wavelet transform : is one of the popular features extraction techniques

###### USing wavelet transform and a raw pixel image we will create our X and use class labels as y. These X and y will be used for model training.

###### wavelet transform  : input ( cropped image) ---> output (transformed image ) 

###### In wavelet transformed image, you can see edges clearly and that can give us clues on various facial features such as eyes, nose, lips etc

### <font color=blue>example for 1 image and after this we will apply on all our dataset</font>

```python
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;   #imArray = imArray / 5
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
cropped_img=cv2.imread('./dataset/cropped/lionel_messi/lionel_messi5.png')
im_har = w2d(cropped_img,'db1',5)
plt.imshow(im_har, cmap='gray')
```

###### Images in cropped folder can be used for model training. We will use these raw images along with wavelet transformed images to train our classifier. Let's prepare X and y now

### <font color=blue> We will use these raw images along with wavelet transformed images to train our classifier. apply on all our dataset and create  X, Y </font>

```python
#ML understand only numeric data so we will assign for each celebrity name a number 
class_dict = {}
count = 0
for celebrity_name in celebrity_file_names_dict.keys():
    class_dict[celebrity_name] = count
    count = count + 1
class_dict
```

```python
X, y = [], []

for celebrity_name, training_files in celebrity_file_names_dict.items():
#for all cropped image 
    for training_image in training_files:
        #read image
        img = cv2.imread(training_image)
        if img is None :
            continue
        #resize image 
        scalled_raw_img = cv2.resize(img, (32, 32))
        #create the  wavelet transformed image,( our function)
        img_har = w2d(img,'db1',5)
        #resize wavelet transformed image
        scalled_img_har = cv2.resize(img_har, (32, 32))
        #create a new image the comination of 2 image scalled_img_har and  scalled_raw_img
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        # define x ,y   ( features , target)
        # x is the combined_img
        #finaly we create our training X 
        X.append(combined_img)
        #y is celebrity_name 
        y.append(class_dict[celebrity_name])
```

# <font color=green>Now we are ready to train our model</font>

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### <font color=blue>GridSearch :Goal is to come up with best modle with best fine tuned parameters</font>

```python
#gridsearchCV run our data for many algo and detect the best algo and the best parameter (hyper parameter)
#define algo and parameters
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [1,10,100,1000],
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,4,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,4,10]
        }
    }
    }
```

### <font color=blue>fit with different algo</font>

```python
from sklearn.pipeline import make_pipeline
scores = []
best_estimators = {}
import pandas as pd
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp['model'])
    clf =  GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model': algo,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

###### we assume that logistic_regression give us the best score so this is the best estimator 

```python
best_clf = best_estimators['logistic_regression']
```

###### confusion_matrix to detect where is some confusion between different class 

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, best_clf.predict(X_test))
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
